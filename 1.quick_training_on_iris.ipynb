```python
from sklearn.ensemble import RandomForestClassifier
```

```python
data = pd.read_excel("iris-train.xlsx")
```

```python
x = data.iloc[:,:-1]
```

```python
x
```

```python
y= data.iloc[:,-1]
```

```python
model = RandomForestClassifier()
```

```python
model.fit(x.values,y.values)
```

```python
predictions = model.predict([[5.4,3.9,1.3,0.4]])
```
